```python
plt.figure(figsize = (10,6))

plt.title('Big O Notation')

plt.xlabel('Elements')
plt.ylabel('Operations')

kwargs = dict(alpha = .6)

x = range(1, 11)




y1 = [(n) + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n ** n for n in x]

plt.plot(x, y1, ls = '-', marker = '*', color = 'orange', label = '$O(1)$', **kwargs)
plt.plot(x, y2, ls = '-', marker = '*', color = 'blue', label = '$Olog(n)$', **kwargs)
plt.plot(x, y3, ls = '-', marker = '*', color = 'green', label = '$O(n)$', **kwargs)
plt.plot(x, y4, ls = '-', marker = '*', color = 'brown', label = '$O(n log n)$', **kwargs)
plt.plot(x, y5, ls = '-', marker = '*', color = 'pink', label = '$O(n^2)$', **kwargs)
plt.plot(x, y6, ls = '-', marker = '*', color = 'purple', label = '$O(2^n)$', **kwargs)
plt.plot(x, y7, ls = '-', marker = '*', color = 'black', label = '$O(n!)$', **kwargs)
plt.plot(x, y8, ls = '-', marker = '*', color = 'red', label = '$O(n^n)$', **kwargs)


plt.xlim(0, 12)
plt.ylim(0, 25)
plt.legend()
plt.tight_layout()
plt.show()
```
