```python
# 1 Use matplotlib to plot the following equation: y = x^2 - x +2

x = range(-12, 12)
y = [n ** 2 - (n + 2) for n in x]

plt.scatter(x, y)

plt.title('$y = x^2 - x + 2$', fontsize = 14)
plt.grid(True, ls = '--')

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2$', fontsize = 14)

plt.xticks([-15, -10, -5, 0, 5, 10, 15], rotation = 45)

plt.annotate('origin', xy = (0, 0), xytext = (0,0), arrowprops = dict(facecolor = 'gray'))
plt.show()
```

```python
#2 Create and label 4 separate charts for the following equations (choose a range for x that makes sense): \sqrt{x}

x = range(0, 50)
y = [n **(1/2) for n in x]

plt.scatter(x, y)

plt.title('$y = \sqrt{x}$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$\sqrt{x}$', fontsize = 14)

plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50  ], rotation = 45)


plt.show()
```

```python
# y = x^3

x = range(-15, 15)

y = [ n ** 3 for n in x]

plt.scatter(x, y)

plt.title('$y = x^3$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^3$', fontsize = 14)

plt.xticks([-20, -15, -10, -5, 0, 5, 10, 15, 20], rotation = 45)

plt.show()
```

```python
# y = 2^x

x = range(-5, 5)

y = [ 2 ** n for n in x]

plt.scatter(x, y)

plt.title('$y = 2^x$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$2^x$', fontsize = 14)

plt.xticks([-7, -5, -3, -1, 0, 1, 3, 5, 7], rotation = 45)

plt.show()
```

```python
# y= 1/(x+1)

x = range(15)
y = [1 /(n + 2) for n in x]

plt.scatter(x, y)

plt.title('$y = 1 /(x+1)$', fontsize = 14)
plt.grid(True, ls = '--')

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2$', fontsize = 14)

plt.xticks([ 0, 10, 20, 30], rotation = 45)

plt.annotate('origin', xy = (0, 0), xytext = (0,0), arrowprops = dict(facecolor = 'gray'))
plt.show()
```

```python
# 3 Combine the figures you created in the last step into one large figure with 4 subplots.

x1 = range(0, 50)
x2 = range(-15, 15)
x3 = range(-5, 5)
x4 = range(15)

y1 = [n **(1/2) for n in x1]
y2 = [ n ** 3 for n in x2]
y3 = [ 2 ** n for n in x3]
y4 = [1 /(n + 2) for n in x4]




```

```python
# title of large graph and size
plt.figure(figsize = (14, 12))

plt.suptitle(' Math Functions', fontsize = 16)


#subplot graph 1
plt.subplot(2,2,1)
plt.plot(x1, y1, ls = '-', marker = '*', color = 'blue')
plt.title('$y = \sqrt{x}$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$\sqrt{x}$', fontsize = 14)

plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50  ], rotation = 45)


#subplot graph 2

plt.subplot(2,2,2)
plt.plot(x2, y2, ls = '-', marker = '*', color = 'pink')
plt.title('$y = x^3$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^3$', fontsize = 14)

plt.xticks([-20, -15, -10, -5, 0, 5, 10, 15, 20], rotation = 45)




#subplot graph 3

plt.subplot(2,2,3)
plt.plot(x3, y3, marker = 'v', ls = '--', color = 'green')
plt.title('$y = 2^x$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$2^x$', fontsize = 14)

plt.xticks([-5, -3, -1, 0, 1, 3, 5], rotation = 45)

#subplot graph 4

plt.subplot(2,2,4)
plt.plot(x4, y4)
plt.title('$y = 1 /(x + 1)$', fontsize = 14)
plt.grid(True, ls = '--')

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2$', fontsize = 14)

plt.xticks([ 0, 10, 20, 30], rotation = 45)

plt.annotate('origin', xy = (0, 0), xytext = (0,0), arrowprops = dict(facecolor = 'gray'))


plt.tight_layout()
plt.show()
```

```python
# 4 Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
plt.figure(figsize = (10,6))

x1 = range(0, 50)
x2 = range(-15, 15)
x3 = range(-5, 5)
x4 = range(15)

y1 = [n **(1/2) for n in x1]
y2 = [ n ** 3 for n in x2]
y3 = [ 2 ** n for n in x3]
y4 = [1 /(n + 2) for n in x4]

plt.title('Combined Functions')

kwargs = dict(alpha = .6)

plt.plot(x1, y1, ls = '-', marker = '*', color = 'orange', label = '$y =\sqrt{x}$', **kwargs)
plt.plot(x2, y2, ls = '-', marker = '*', color = 'blue', label = '$y = x^3$', **kwargs)
plt.plot(x3, y3, ls = '-', marker = '*', color = 'green', label = '$y = 2^x$', **kwargs)
plt.plot(x4, y4, ls = '-', marker = '*', color = 'brown', label = '$y = 1 /(x + 1)$', **kwargs)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$y$', fontsize =14)
plt.xlim(-5, 15)
plt.ylim(-5, 15)

plt.legend()
plt.tight_layout()
plt.show()
```
